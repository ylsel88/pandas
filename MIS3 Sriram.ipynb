```python
import pandas as pd
import xml.etree.ElementTree as et

#Load XML file
xml_file = "FLOW_GPF_HK.20190211.PNL.xml"
xtree = et.parse(xml_file)
xroot = xtree.getroot()

#Include fields at root level
df_cols = ["cobDate","signOffGroup","riskBook","underlyingRiskBook","tradeBook","tradeId","Enrichment_version","riskTrade","reportName","value","riskCurrency","riskId","valueCcy","reportClass","subReportClass"]
#Add columns to be extracted 
out_df = pd.DataFrame(columns = df_cols)

#Include fields at root level
s_cobDate = xroot.find("cobDate").text if  xroot.find("cobDate") is not None else None
s_signOffGroup = xroot.find("signOffGroup").text if xroot.find("signOffGroup") is not None else None
for book in xroot.iter('book'):
    #Include book level fields
	s_tradeBook = book.find("tradeBook").text if book.find("tradeBook") is not None else None
	s_riskBook = book.find("riskBook").text if book.find("riskBook") is not None else None
	s_underlyingRiskBook = book.find("underlyingRiskBook").text if  book.find("underlyingRiskBook") is not None else None
	for trade in book.iter('trade'):
        #Include trade level fields
		s_tradeId = trade.find("tradeId").text if trade.find("tradeId") is not None else None
		enrichment = trade.find('enrichment')
		s_en_version = enrichment.find("version").text if enrichment.find("version") is not None else None 
		for subtrade in trade.iter('subtrade'):
            #Include subtrade level fields
			s_riskTrade = subtrade.find("riskTrade").text if subtrade.find("riskTrade") is not None else None  
			for report in subtrade.iter('report'):
                #Include report fields
				s_reportName = report.find("reportName").text if report.find("reportName") is not None else None
				s_riskCurrency = report.find("riskCurrency").text if report.find("riskCurrency") is not None else None
				s_riskId = report.find("riskId").text if report.find("riskId") is not None else None
				s_reportClass = report.find("reportClass").text if report.find("reportClass") is not None else None
				s_subReportClass = report.find("subReportClass").text if report.find("subReportClass") is not None else None
				for value in report.iter('value'):
					#Include value fields
					s_value = value.find("value").text if value.find("value") is not None else None
					s_valueCcy = value.find("valueCcy").text if value.find("valueCcy") is not None else None
					out_df = out_df.append(pd.Series([s_cobDate,s_signOffGroup,s_riskBook,s_underlyingRiskBook,s_tradeBook,s_tradeId,s_en_version,s_riskTrade,s_reportName,s_value,s_riskCurrency,s_riskId,s_valueCcy,s_reportClass,s_subReportClass], 
                                      index = df_cols), ignore_index = True)
```
